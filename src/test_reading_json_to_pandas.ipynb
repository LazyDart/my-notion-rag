```python
def remove_needless_data(df):
    df = df.drop(
        ["object", "created_by", "last_edited_by", 
         "cover", "icon", "parent", 
         "in_trash", "url", "public_url"
         "archived"],
        axis=1)

    return df
```

```python
def extract_properties(df):
    df = df.copy()

    extracted_properties = list(zip(*df["properties"].apply(lambda x: list(x.values()))))
    
    for i, key in enumerate(df["properties"].iloc[0].keys()):
        df[key] = extracted_properties[i]

    return df, list(df["properties"].iloc[0].keys())
```

```python
notes, prop_names = extract_properties(notes)
```

```python
# TODO use cases?
def process_property(prop_dict):
    if prop_dict["type"] == "status":
        return prop_dict["status"]["color"]  # Status indication
    
    elif prop_dict["type"] == "formula":
        form_type = prop_dict["formula"]["type"]  # Types of formulas can differ
        return prop_dict["formula"][form_type]  
    
    elif prop_dict["type"] == "relation":
        return [rel_dict["id"] for rel_dict in prop_dict["relation"]]  # Extract all related pages (projects/areas)
    
    elif prop_dict["type"] == "multi_select":
        return [selection["name"] for selection in prop_dict["multi_select"]] # All choosen selections
    
    elif prop_dict["type"] == "checkbox":
        return prop_dict["checkbox"]  # True/False
    
    elif prop_dict["type"] == "created_time":
        return prop_dict["created_time"]
    
    elif prop_dict["type"] == "url":
        return prop_dict["url"]  # Str Link or None
    
    elif prop_dict["type"] == "title":
        return prop_dict["title"][0]["text"]["content"]  # Text
    
    else:
        return prop_dict
```

```python
for col in prop_names:
    notes[col] = notes[col].apply(process_property)
    
```

```python
projects = remove_needless_data(projects)
areas = remove_needless_data(areas)

projects, prop_names = extract_properties(projects)

for col in prop_names:
    projects[col] = projects[col].apply(process_property)

areas, prop_names = extract_properties(areas)

for col in prop_names:
    areas[col] = areas[col].apply(process_property)
```
