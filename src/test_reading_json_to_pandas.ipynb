```python
def remove_needless_data(df):
    df = df.drop(
        ["object", "created_by", "last_edited_by", 
         "cover", "icon", "parent", 
         "in_trash", "url", "public_url"
         "archived"],
        axis=1)

    return df
```

```python
def extract_properties(df):
    df = df.copy()

    extracted_properties = list(zip(*df["properties"].apply(lambda x: list(x.values()))))
    
    for i, key in enumerate(df["properties"].iloc[0].keys()):
        df[key] = extracted_properties[i]

    return df, list(df["properties"].iloc[0].keys())
```
